```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
fx = x ** 3 - 50 * x
gx = -x ** 4 + 88 * x ** 2 - 241
```

```python
inters = solve(fx - gx, x)
```

```python
inters_x = []
```

```python
for i in inters:
    z = round(i.n(), 2)
    inters_x.append(z)
```

```python
inters_y = []
```

```python
for i in inters_x:
    y1 = fx.subs(x, i)
    inters_y.append(y1)
```

```python
inters_y
```

```python
for i, (x, y) in enumerate(zip(inters_x, inters_y), start = 1):
    print(f'точка пересечения {i} : ({x},{y})')
```
